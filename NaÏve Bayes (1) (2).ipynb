```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(['id','diagnosis','Unnamed: 32'],axis=1),df['diagnosis'],test_size=0.2)
```

```python
class NaiveBayes:
    def __init__(self,df):
        self.df=df
        self.maindict=maindict={}
        self.pred=pred=[]
        self.pre=pre={}
    def train(self,X_train,y_train):
        self.X_train=X_train
        X_train=self.X_train
        X_train['preds']=y_train
        X_trainordered=X_train.reset_index(inplace=False)
        maindict=self.maindict
        pre=self.pre
        for uniq in X_trainordered['preds'].unique():
            dicts={}
            df1=X_trainordered[X_trainordered['preds']==uniq]
            df1.drop(['index','preds'],inplace=True,axis=1)
            df1.reset_index(inplace=True)
            df1.drop('index',inplace=True,axis=1)
            pre[uniq]=len(df1)/len(X_train)
            print(pre[uniq])
            for columns in df1.columns:
                mean=(df1[columns].sum()/len(df1))
                errorsum=0
                for rows in range(0,len(df1)):
                    errorsum=errorsum+(df1[columns][rows]-mean)**2
                stdev=((errorsum/(len(df1)-1))**0.5)
                dicts[columns]=[mean,stdev]
            maindict[uniq]=dicts
    def test(self,X_test):
        self.X_test=X_test
        X_test=self.X_test
        X_train=self.X_train
        maindict=self.maindict
        pre=self.pre
        pred=self.pred
        X_testordered=X_test.reset_index()
        prob={}
        predictions=[]
        for rows in range(0,len(X_test)):
            for uniq in X_train['preds'].unique():
                p=1
                for columns in X_test.columns:
                    mean=maindict[uniq][columns][0]
                    stdev=maindict[uniq][columns][1]
                    p=p*(1/math.sqrt(2*3.14*stdev))*math.exp(-0.5*math.pow(((X_testordered[columns][rows]-mean)/stdev),2))
                pt=p*pre[uniq]
                prob[uniq]=pt
            print(prob)
            pred.append(max(zip(prob.values(),prob.keys()))[1])                               
        return pred      
    def accuracy(self,y_test):
        self.y_test=y_test
        a=np.array(self.y_test)
        correctcnt=0
        total=len(a)
        for i in range(0,len(a)):
            if a[i]==pred[i]:
                correctcnt=correctcnt+1
        accuracy=correctcnt/total    
        print(accuracy)
        
        
    
```
